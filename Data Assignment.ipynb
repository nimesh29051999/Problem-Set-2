```python
'''
Ques3
'''
class Marsupial:
    def __init__(self):
        self.pouch = []
    def put_in_pouch(self, element):
        self.pouch.append(element)
    def pouch_contents(self):
        print(self.pouch)
        
class Kangaroo(Marsupial):
    def __init__(self, x, y):
        Marsupial.__init__(self);
        self.x = x
        self.y = y
    def jump(self, x, y):
        self.x += x
        self.y += y
    def __str__(self):
        return 'I am a Kangaroo located at coordinates({},{})'.format(self.x, self.y)
```

```python
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
m.pouch_contents()

k = Kangaroo(0,0)
print(k)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
k.pouch_contents()
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)
```

```python
from bs4 import BeautifulSoup
import requests

urls = []
indent = 0
def scrape(site, depth, level):
    if level >= depth:
        return
    global indent
    r = requests.get(site)
    s = BeautifulSoup(r.text,"html.parser")
    for i in s.find_all("a"):
        href = None
        if 'href' in i.attrs:
            href = i.attrs['href']
        if href and href.startswith("/"):
            site = site+href
            if site not in urls:
                urls.append(site)
                for j in range(indent):
                    print("\t",end='')
                # increase the indent
                indent += 1
                print(indent,site[:30])
                # call the child site
                scrape(site, depth, level+1)
                # decrease indent
                indent -= 1
                
def webdir(site, depth, indent):
    scrape(site, depth, 0)
```

```python
webdir('https://en.wikipedia.org/wiki/Main_Page', 4, 0)
```
